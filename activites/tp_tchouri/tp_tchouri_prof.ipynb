### Pour fonctionner, ce notebook doit être dans le même dossier que le module CelestialBody !

```python
%pylab inline
from celestialbody import CelestialBody
from datetime import datetime, timedelta
import astropy.constants as const
```

# Création des fichiers de données

```python
for name, step in zip(["Earth", "Jupiter", "Churyumov-Gerasimenko"], [14, 70, 49]):
    body = CelestialBody(name)

    start = datetime.today()
    stop  = start + timedelta(days = body.period)
    next_date = start+timedelta(days=step)

    header = [u"# fichier "+body.name.lower()+".txt\n",
              u"############################################################################################\n",
              u"# Positions prévues de "+body.name+" par rapport au Soleil.\n",
              u"# Les données sont prises tous les %d jours.\n"%(step),
              u"# La première ligne correspond au %2d/%2d/%4d, la deuxième au %2d/%2d/%4d, etc.\n"%(start.day, start.month, start.year, next_date.day, next_date.month, next_date.year),
              u"# Les distances sont exprimées en unité astronomique (au) : 1 au = %9.3e m\n"%(const.au.value),                    u"############################################################################################\n"
             ]

    body.data_position_txt(start=start, stop=stop, step=step, cols="xy", header=header)
```

Le fichier churyumov-gerasimenko.txt doit être renommé ensuite en tchouri.txt pour correspondre à l'activité proposée, ou alors il faudra adapter le programme élève comme cela est fait dans l'exemple ci-dessous.

# Un programme élève complet

```python
import numpy as np
import matplotlib.pyplot as plt

###########################################################################
# LECTURE DES FICHIERS DE DONNÉES POUR RÉCUPÉRER LES POSITIONS DES PLANÈTES
###########################################################################
# Terre
data_terre = np.loadtxt("earth.txt", skiprows=8)
x_terre, y_terre = data_terre[:,0], data_terre[:,1]
# Tchouri
data_tchouri = np.loadtxt("churyumov-gerasimenko.txt", skiprows=8)
x_tchouri, y_tchouri = data_tchouri[:,0], data_tchouri[:,1]
# Jupiter
data_jupiter = np.loadtxt("jupiter.txt", skiprows=8)
x_jupiter, y_jupiter = data_jupiter[:,0], data_jupiter[:,1]

##############################################################
# REPRESENTATION GRAPHIQUE DES POSITIONS DES DIFFÉRENTS OBJETS
##############################################################
fig = plt.figure(figsize=(8,8))

# Le Soleil
plt.plot([0],[0], "*", color="gold", markersize=20, label="Soleil")

# TERRE
plt.plot(x_terre, y_terre, "o", color="blue", label="Terre")

# TCHOURI
plt.plot(x_tchouri, y_tchouri, "o", color="black", label="Tchouri")

# JUPITER
plt.plot(x_jupiter, y_jupiter, "o", color="red", label="Jupiter")

# Changement d'échelle pour les axes (modification des limites du graphe)
plt.xlim(-6,6)
plt.ylim(-6,6)

# Titre, nom des axes et légende
plt.title("Évolution de la position de quelques objets du système solaire")
plt.xlabel("Abscisse par rapport au soleil\nen unités astronomiques (au)")
plt.ylabel("Ordonnée par rapport au soleil\nen unités astronomiques (au)")
plt.legend(loc = "upper left")
fig.axes[0].set_aspect("equal")
plt.show()
```
